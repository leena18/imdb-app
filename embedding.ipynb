```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'Understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
#Define the vocab size and convert into verctor
vac_size = 10000
```

```python
# one Hot Representation
one_hot_repr = [one_hot(words, vac_size)for words in sent]
one_hot_repr
```

```python

from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sen_len = 8
embedded_doc = pad_sequences(one_hot_repr,padding='pre',maxlen=sen_len)
print(embedded_doc)
```

```python
## Feature representation 
dim = 10
```

```python
model = Sequential()
model.add(Embedding(vac_size,dim,input_length=sen_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
